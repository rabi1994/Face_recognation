```python
import albumentations as A
```

```python
transoforms =A.Compose([A.IAASharpen(always_apply=True,p=1)])
```

```python
transform =transoforms(image=read_image(training[52][0]))
```

```python
plt.figure(figsize=(20,9))
plt.axis('off')
plt.imshow(transform['image'])
```

```python
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
image = transform['image']
gray_image = cv2.cvtColor(transform['image'],cv2.COLOR_RGB2GRAY)
frame_gray = cv2.equalizeHist(gray_image)
```

```python
plt.imshow(frame_gray,cmap='gray')
```

```python
faces = face_cascade.detectMultiScale(frame_gray, 1.1, 4)
for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)
plt.figure(figsize=(14,8))
plt.axis('off')
plt.imshow(image)
```

### trying all the augmantations and see what will work and what will destroy our data 

```python
all_augmantations ={"Gaussian blur":A.Compose([A.GaussianBlur(blur_limit=(13,23),sigma_limit=0,always_apply=True,p=1)]),
                   "Median blur":A.Compose([A.MedianBlur(blur_limit=47,always_apply=True,p=1)]),
                   "Horizontal flip":A.Compose([A.HorizontalFlip(always_apply=True,p=1)]),
                   "Vertical flip":A.Compose([A.VerticalFlip(always_apply=True,p=1)]),
                   "Rotate": A.Compose([A.Rotate(border_mode=1,limit=45,p=1,always_apply=True)]),
                   "Random crop": A.Compose([A.RandomCrop(height=300,width=300,always_apply=True)]),
                   "Zoom": A.Compose([A.CenterCrop(height=300,width=300,always_apply=True)]),
                   "sharpen":A.Compose([A.IAASharpen(always_apply=True)]),
                   "Random rain": A.Compose([A.RandomRain(always_apply=True)]),
                   "Random Fog":A.Compose([A.RandomFog(always_apply=True)])}
```

```python
i=1
augmanted_images=[]
plt.figure(figsize=(30,12))
for augmantation in list(all_augmantations.keys()):
    #print(augmantation)
    #print(all_augmantations[augmantation])
    transforms =all_augmantations[augmantation]
    transform =transforms(image=read_image(training[52][0]))
    plt.subplot(4,3,i)
    i +=1
    plt.title(augmantation)
    plt.axis('off')
    plt.imshow(transform['image'])
    augmanted_images.append(transform['image'])
```

```python
%time list(all_augmantations.keys())
```

```python
plt.figure(figsize=(20,9))
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
i=1
for augmanted_image in augmanted_images:
    image = augmanted_image
    gray_image = cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)
    frame_gray = cv2.equalizeHist(gray_image)
    faces = face_cascade.detectMultiScale(frame_gray, 1.1, 4)
    for (x, y, w, h) in faces:
        cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)
    plt.subplot(4,3,i)
    i += 1
    plt.axis('off')
    plt.imshow(image)
```
